```python
import os
import random
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
DATA_DIR = Path("../data")
ORIGINAL_DATA_SUBDIR = Path("Pepsico RnD Potato Lab Dataset")
PREPROCESSED_DATA_SUBDIR = Path("Preprocessed Dataset")
TRAIN_SUBDIR = Path("Train")
TEST_SUBDIR = Path("Test")
```

```python
def display_image(image, grayscale=False):
    plt.figure()
    
    if grayscale:
        plt.imshow(image, cmap="gray")
    else:
        plt.imshow(image)
    
    return
```

```python
classes = os.listdir(DATA_DIR / ORIGINAL_DATA_SUBDIR / TRAIN_SUBDIR)
```

```python
train_filepaths_defective = [DATA_DIR / ORIGINAL_DATA_SUBDIR /TRAIN_SUBDIR / classes[1] / file for file in os.listdir(DATA_DIR / ORIGINAL_DATA_SUBDIR / TRAIN_SUBDIR / classes[1])]
train_filepaths_nondefective = [DATA_DIR / ORIGINAL_DATA_SUBDIR /TRAIN_SUBDIR / classes[0] / file for file in os.listdir(DATA_DIR / ORIGINAL_DATA_SUBDIR / TRAIN_SUBDIR / classes[0])]
test_filepaths_defective = [DATA_DIR / ORIGINAL_DATA_SUBDIR / TEST_SUBDIR / classes[1] / file for file in os.listdir(DATA_DIR / ORIGINAL_DATA_SUBDIR / TEST_SUBDIR / classes[1])]
test_filepaths_nondefective = [DATA_DIR / ORIGINAL_DATA_SUBDIR / TEST_SUBDIR / classes[0] / file for file in os.listdir(DATA_DIR / ORIGINAL_DATA_SUBDIR / TEST_SUBDIR / classes[0])]
```

```python
plt.figure(figsize=(10, 10))

for i, file in enumerate(random.sample(train_filepaths_defective, 9)):
    ax = plt.subplot(3, 3, i + 1)
    image = plt.imread(file)
    plt.imshow(image)
    plt.title(str(file).split("/")[-1])
    plt.axis("off")
```

```python
plt.figure(figsize=(10, 10))

for i, file in enumerate(random.sample(train_filepaths_nondefective, 9)):
    ax = plt.subplot(3, 3, i + 1)
    image = plt.imread(file)
    plt.imshow(image)
    plt.title(str(file).split("/")[-1])
    plt.axis("off")
```
