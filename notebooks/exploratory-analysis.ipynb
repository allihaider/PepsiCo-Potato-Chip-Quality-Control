```python
import os
import random
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
CLASSES = ["Defective", "Non-Defective"]

DATA_DIR = Path("../data")
ORIGINAL_DATA_SUBDIR = Path("Pepsico RnD Potato Lab Dataset")
PREPROCESSED_DATA_SUBDIR = Path("Preprocessed Dataset")
TRAIN_SUBDIR = Path("Train")
TEST_SUBDIR = Path("Test")
```

```python
def display_image(image, grayscale=False):
    plt.figure()
    
    if grayscale:
        plt.imshow(image, cmap="gray")
    else:
        plt.imshow(image)
    
    return
```

```python
train_filepaths_defective_og = [DATA_DIR / ORIGINAL_DATA_SUBDIR / TRAIN_SUBDIR / CLASSES[0] / file for file in os.listdir(DATA_DIR / ORIGINAL_DATA_SUBDIR / TRAIN_SUBDIR / CLASSES[0])]
train_filepaths_nondefective_og = [DATA_DIR / ORIGINAL_DATA_SUBDIR / TRAIN_SUBDIR / CLASSES[1] / file for file in os.listdir(DATA_DIR / ORIGINAL_DATA_SUBDIR / TRAIN_SUBDIR / CLASSES[1])]
train_filepaths_defective_processed = [DATA_DIR / PREPROCESSED_DATA_SUBDIR / TRAIN_SUBDIR / CLASSES[0] / file for file in os.listdir(DATA_DIR / PREPROCESSED_DATA_SUBDIR / TRAIN_SUBDIR / CLASSES[0])]
train_filepaths_nondefective_processed = [DATA_DIR / PREPROCESSED_DATA_SUBDIR / TRAIN_SUBDIR / CLASSES[1] / file for file in os.listdir(DATA_DIR / PREPROCESSED_DATA_SUBDIR / TRAIN_SUBDIR / CLASSES[1])]
```

```python
plt.figure(figsize=(10, 10))

for i, file in enumerate(random.sample(train_filepaths_defective_og, 9)):
    ax = plt.subplot(3, 3, i + 1)
    image = plt.imread(file)
    plt.imshow(image)
    plt.title(str(file).split("/")[-1])
    plt.axis("off")
```

```python
plt.figure(figsize=(10, 10))

for i, file in enumerate(random.sample(train_filepaths_defective_processed, 9)):
    ax = plt.subplot(3, 3, i + 1)
    image = plt.imread(file)
    plt.imshow(image)
    plt.title(str(file).split("/")[-1])
    plt.axis("off")
```
